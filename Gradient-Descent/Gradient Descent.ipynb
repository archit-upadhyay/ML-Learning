```python
## Gradient Descent Algo - Question 1 (Trial)
import numpy as np
import matplotlib.pyplot as plt

def cost_func(x):
  return ((x**2)-(4*x)+3)

def grad_slope(x):
  return (2*x) - 4


x = np.random.randint(100)
print("Value of x:",x)
alpha = 0.1
num_iters = 100
cost_values = []
x_values = []

for i in range(num_iters):
  #print(i, "iteration")
  cost = cost_func(x)
  #print("Cost",cost)
  grad = grad_slope(x)
  #print("Grad",grad)

  x = x - alpha * grad
  #print("Updated x:",x)
  #print(cost)
  #print('\n')
  x_values.append(x)
  cost_values.append(cost)

print(cost_values)
print(x_values)

plt.plot(cost_values)
plt.plot(x_values)



```
